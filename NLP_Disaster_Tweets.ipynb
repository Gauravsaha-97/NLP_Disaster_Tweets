```
# Importing Libraries and Load the Datasets
import tensorflow as tf
import numpy as np
import pandas as pd

test_data = pd.read_csv("/content/drive/My Drive/Tweets_Disaster/Tweet_Disaster/test (1).csv")
train_data = pd.read_csv("/content/drive/My Drive/Tweets_Disaster/Tweet_Disaster/train.csv")
sample_sub = pd.read_csv("/content/drive/My Drive/Tweets_Disaster/Tweet_Disaster/sample_submission.csv")
```

```
# Show the actual Training Dataset
train_data.head()
```

```
# Shape of training data
train_data.shape
```

```
# Drop the unimportant columns in Training Data
cols_to_drop = ['id','keyword','location']
df_train = train_data.drop(cols_to_drop, axis=1)
df_train.head()
```

```
# First 5 rows of Test data 
test_data.head()
```

```
# Drop the columns in Test Data
cols_to_drop = ['id','keyword','location']
df_test = test_data.drop(cols_to_drop, axis=1)
df_test.head()
```

```
# Get the list of sentences to train the model
train_sentences = list(df_train['text'])
train_sentences
```

```
# List of Training Labels
train_labels = list(df_train['target'])
train_labels
```

```
# Get the list of Testing sentences to predict
validation_sentences = list(df_test['text'])
print(len(validation_sentences))
```

```
from tensorflow.keras.preprocessing.text import Tokenizer
```

```
# Setting hyperparameters and tokenizing the training sentences
num_words = 2000
oov_tok = "<OOV>"
max_length = 80
pad_type='post'
embed_dim = 20
tokenizer = Tokenizer(num_words=num_words, oov_token=oov_tok)
tokenizer.fit_on_texts(train_sentences)
word_index = tokenizer.word_index
print(word_index)
```

```
training_sentence_seq = tokenizer.texts_to_sequences(train_sentences)
training_sentence_seq
```

```
# Create the padding matrix to fed the neural network
from tensorflow.keras.preprocessing.sequence import pad_sequences
training_sentences_pad = pad_sequences(training_sentence_seq, maxlen=max_length, padding=pad_type)
training_sentences_pad
```

```
training_labels = np.array(train_labels).reshape((-1,1))
print(training_sentences_pad.shape, training_labels.shape)
```

```
validation_sentences_seq = tokenizer.texts_to_sequences(validation_sentences)
validation_sentences_pad = pad_sequences(validation_sentences_seq, padding=pad_type, maxlen=max_length)

print(validation_sentences_pad.shape)
```

```
# Define the model
model = tf.keras.models.Sequential([
                                    tf.keras.layers.Embedding(input_dim=num_words,output_dim=embed_dim, input_length=max_length),
                                    tf.keras.layers.GlobalAveragePooling1D(),
                                    tf.keras.layers.Dense(16, activation='relu'),
                                    tf.keras.layers.Dense(1,activation='sigmoid')
])
```

```
# Compile the model
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.summary()
```

```
# Train the model
num_epochs=30
model.fit(training_sentences_pad, training_labels, epochs=num_epochs)
```

##### Training Accuracy after 30 epochs is 88.9%

```
# Make predictions
y_pred = model.predict_classes(validation_sentences_pad)
y_pred = pd.DataFrame(y_pred)
```

```
predicted_labels = (model.predict(validation_sentences_pad) > 0.5).astype("int32")
predicted_labels = pd.DataFrame(predicted_labels)
```
